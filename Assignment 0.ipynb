```python
# Q1. You are given a list of n integers. Find and print the maximum and minimum values in the list without
#     using Python's built-in max() or min() functions.

def max_min(numbers): #function for finding maximum and minimum from the list
  if not numbers:
      return None, None

  maximum= numbers[0]
  minimum= numbers[0]

  for num in numbers:
    if num>maximum:
        maximum=num

    if num<minimum:
        minimum=num

  return maximum, minimum
    
```

```python
# Question 3
# You are given a list of n integers and a number x. Find and print any pair of elements from the list
# whose absolute difference equals x. You are guaranteed that such a pair exists.

def difference(n,x):
    if not n: 
        return None

    n1= set()

    for num in n:
        if (num + x) in n1:
            return(num, num+x)

        if (num - x) in n1:
            return(num-x, num)

    
```

```python
import pandas as pd

data = {
    'Ticker': ['RELIANCE', 'TCS', 'INFY', 'HDFCBANK', 'WIPRO'],
    'Price': [2850.5, 3900.0, 1500.75, 1650.0, 420.3],
    'Volume': [1200000, 850000, 1100000, 950000, 700000],
    'Sector': ['Energy', 'IT', 'IT', 'Finance', 'IT']
}

df = pd.DataFrame(data)
df['Market_Cap_Proxy'] = df['Price'] * df['Volume']

# (a) Mean price across all stocks
mean_price = df['Price'].mean()
print(f"(a) Mean Price: {mean_price}")

# (b) Total volume traded in IT sector
it_volume = df[df['Sector'] == 'IT']['Volume'].sum()
print(f"(b) Total IT Volume: {it_volume}")

# (c) Stock with highest Market_Cap_Proxy
top_ticker = df.loc[df['Market_Cap_Proxy'].idxmax(), 'Ticker']
print(f"(c) Highest Market Cap Stock: {top_ticker}")
```

```python
import numpy as np

scores = np.array([45, 82, 76, 55, 90, 38, 67, 71])

# (a) Mean, Median, Standard Deviation
mean   = np.mean(scores)
median = np.median(scores)
std    = np.std(scores)

print("(a) Statistics:")
print(f"    Mean:               {mean}")
print(f"    Median:             {median}")
print(f"    Standard Deviation: {std:.2f}")

# (b) Normalize between 0 and 1
normalized = (scores - scores.min()) / (scores.max() - scores.min())
print(f"\n(b) Normalized Scores: {np.round(normalized, 2)}")

# (c) Indices of students who scored above the mean
above_mean_indices = np.where(scores > mean)[0]
print(f"\n(c) Indices of students above mean ({mean}): {above_mean_indices}")
print(f"    Their scores: {scores[above_mean_indices]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 252 daily returns
np.random.seed(42)
daily_returns = np.random.normal(loc=0.0008, scale=0.015, size=252)

# (a) Construct price series starting at 100
price_series = 100 * np.cumprod(1 + daily_returns)

# (b) Plot the price series
plt.figure(figsize=(12, 6))
plt.plot(price_series, color='steelblue', linewidth=1.5)
plt.title('Simulated Stock Price Series (252 Trading Days)')
plt.xlabel('Trading Day')
plt.ylabel('Price')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# (c) Final price, Peak price, Maximum Drawdown
final_price = price_series[-1]
peak_price  = np.max(price_series)

# Maximum Drawdown
rolling_peak = np.maximum.accumulate(price_series)
drawdowns    = (price_series - rolling_peak) / rolling_peak
max_drawdown = np.min(drawdowns) * 100

print(f"Final Price:      {final_price:.2f}")
print(f"Peak Price:       {peak_price:.2f}")
print(f"Maximum Drawdown: {max_drawdown:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
year1 = [42, 45, 50, 48, 55, 60, 58, 63, 67, 70, 75, 80]
year2 = [52, 49, 58, 61, 66, 71, 68, 74, 79, 83, 88, 95]

difference = [y2 - y1 for y1, y2 in zip(year1, year2)]
colors     = ['green' if d > 0 else 'red' for d in difference]

# Create figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# (a) Line chart — both years
ax1.plot(months, year1, marker='o', color='steelblue',  linewidth=2, label='Year 1')
ax1.plot(months, year2, marker='o', color='darkorange', linewidth=2, label='Year 2')
ax1.set_title('Monthly Revenue Comparison: Year 1 vs Year 2')
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue (in Lakhs)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# (b) Bar chart — month-by-month difference
ax2.bar(months, difference, color=colors, edgecolor='black', linewidth=0.5)
ax2.axhline(y=0, color='black', linewidth=0.8)
ax2.set_title('Month-by-Month Revenue Difference (Year 2 - Year 1)')
ax2.set_xlabel('Month')
ax2.set_ylabel('Difference (in Lakhs)')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'Day':    range(1, 11),
    'Price':  [100, 103, np.nan, 108, 107, np.nan, 112, 115, np.nan, 120],
    'Volume': [500, 520, 510, np.nan, 530, 545, np.nan, 560, 570, 580]
})

# (a) Count missing values in each column
print("(a) Missing values per column:")
print(df.isnull().sum())

# (b) Fill missing values
df['Price']  = df['Price'].ffill()                        # forward fill
df['Volume'] = df['Volume'].fillna(df['Volume'].mean())   # fill with mean

print("\n(b) DataFrame after filling missing values:")
print(df)

# (c) Daily percentage return
df['Return'] = df['Price'].pct_change() * 100

print("\n(c) DataFrame with Return column:")
print(df.round(2))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate price data
np.random.seed(42)
returns = np.random.multivariate_normal(
    mean=[0.0005, 0.0008, 0.0003],
    cov=[[0.0004,  0.0001,  0.00005],
         [0.0001,  0.0006,  0.00008],
         [0.00005, 0.00008, 0.0003]],
    size=200
)
prices = 100 * np.cumprod(1 + returns, axis=0)

# (a) Convert to DataFrame
df = pd.DataFrame(prices, columns=['Asset_A', 'Asset_B', 'Asset_C'])

print("(a) First 5 rows of DataFrame:")
print(df.head().round(2))

# (b) + (d) Plot all three assets and portfolio on same chart
df['Portfolio'] = df[['Asset_A', 'Asset_B', 'Asset_C']].mean(axis=1)

plt.figure(figsize=(12, 6))
plt.plot(df['Asset_A'],  linewidth=1.5, label='Asset A',          color='steelblue')
plt.plot(df['Asset_B'],  linewidth=1.5, label='Asset B',          color='darkorange')
plt.plot(df['Asset_C'],  linewidth=1.5, label='Asset C',          color='green')
plt.plot(df['Portfolio'],linewidth=2.5, label='Equal-Weight Portfolio',
         color='red', linestyle='--')
plt.title('Asset Price Series with Equal-Weight Portfolio (200 Trading Days)')
plt.xlabel('Trading Day')
plt.ylabel('Price')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# (c) Correlation matrix of daily returns
daily_returns = df[['Asset_A', 'Asset_B', 'Asset_C']].pct_change().dropna()
corr_matrix   = daily_returns.corr()

print("\n(c) Correlation Matrix of Daily Returns:")
print(corr_matrix.round(4))
```
